리스트(list)
- 리스트는 목록
-리스트 안에는 아무거나 들어갈수 있음(숫자, 문자 상관없다)


```python
fruits = ['apple','bear','orange']
print(fruits[0])

fruits.append('mongo')
print(fruits)
```

```python
#리스트 생성하기
num = [1,2,3,4,5]
fruits = ['apple','bear','orange']
mixed_list = [10,"hello", True, 3.14]

#리스트 요소에 접근하기
print("첫 번째 과일:", fruits[0]) #apple
print("두 번째 숫자:", num[1]) #2

#리스트 요소 수정하기
fruits[0] = "cherry"
print("수정된 첫 번째 과일:", fruits[0]) #cherry

# 리스트에 요소 추가하기
fruits.append("grape")
print("추가된 과일:", fruits)

# 리스트에서 요소 제거하기
fruits.remove("bear")
print("제거된 과일:", fruits)

```

딕셔너리(dictionary)

```python
a_dict = {} #빈 딕셔너리 생성
print(a_dict)

a_dict = {'name': '이예진','age':25}
print(a_dict)

print(a_dict['name'])
```

```python
# 사전 생성하기
student = {
      "name":"emma",
      "age": 25,
      "sex": "female"
}

#key를 사용하여 값에 접근하기
print("name: ", student["name"])
print("age: ", student["age"])

#사전 값 수정하기
student["age"] = 26
print("new age: ", student["age"])

#새로운 키-값 쌍 추가하기
student["school"] = "korea univercity"
print("school: ", student["school"])

# 키를 사용하여 사전에서 항목 제거하기
del student["sex"]
print("remove sex: ", student)
```

특정 문자를 기준으로 문자열 나누기

```python
#이메일 주소에서 도메인 'gmail' 만 추출하기

myemail = 'test@gmail.com'

print(myemail)
print(myemail.split('@'))

result = myemail.split('@')
print('--------------------------')

print(result[0])
print(result[1])

```

```python
result2 = result[1].split('.')
print(result2)
```

```python
print(result2[0]) #gmail
print(result2[1]) #com
```

```python
#한 줄로 한번에 처리 (method chaining 메서드 체인)
myemail.split('@')[1].split('.')[0]
```

질의응답

```python
'''
이메일이나 주소를 사용하다 보면 옛날에 co.kr 이런거 쓰던경우가 있었는데
이때 .을 기준으로 분리하면 co 와 kr 도 분리되어 버리는데 이부분은 분리 안되게 할려면 어떻게 해야하나요?
'''
```

```python
mail = 'yahoo.co.kr'
a = mail.split(".")
print(a)

b = '.'.join([a[1],a[2]])
print(b)
```

```python
mail.split('.',maxsplit = 1)
```
