```python
mycampus ='나는 현재 대한민국 최고의 IT 데이터분석 & 개발자 양성소인 멀티 캠퍼스 교육생이다.\
억대 연봉자.데이터사이어티스가 되겠다는 나의 꿈을 바로 이 곳, 멀티 캠퍼스에서 이룰 것이다.'

posList = [ ]
position = 0

while True:
    try:
        multi = mycampus.index('멀티',position)
        posList.append(multi)
        position += 1
    except:
        break

print("멀티 글자 위치: ", sorted(list(set(posList))))



```
