Tuple(튜플) 정의

```python
t1 = ()
t2 = (1,2,3,4,5)
t3 = (2,4,'판다','강아지') #컬렉션 타입(여러 개의 자료형 소유가능)
```

```python
t2 = (1,2,3,4,5)
print(t2[1:2])
print(t2[1])
```

```python
#인덱싱(Indexing)
print(t2[0])
print(t2[-1])

#인덱싱을 통해 튜플 내부 데이터 접근 가능
#파이썬에서 index는 0부터 시작
#파이썬에서 index에 '-'기호가 붙으면, 뒤에서부터 숫자를 센다.(-1,-2,-3....)

print()
print(t3[-2])
print(t3[-3])
```

```python
#슬라이싱(slicing)
#>> 튜플 안의 값의 일부만 가져오는 것
t2 = (1,2,3,4,5)
print(t2[2:])
print(t2[:4]) #n-1개까지
print(t2[1:3])
print(t2[::2]) #간격(interval)이 2칸씩 이동
print(t2[::-1]) #역순으로 -1칸씩 이동
```

교집합, 합집합, 차집합

```python
t2 = set([1,2,3,4,5])
t3 = set([2,4,'판다','강아지'])
```

```python
#교집합 intersection()

print(t2.intersection(t3))
print( t2 & t3)
```

```python
#차집합 difference
print(t2.difference(t3))
print( t2 - t3)
```

```python
#합집합 union
print(t2.union(t3))
print( t2 | t3)
```
