```python
#while문 활용 menu 주문 프로그램만들기
#1유형

menu = {
    '떡볶이' : '순대',
    '짜장면' : '짬뽕',
    '라면' : '김치',
    '피자' : '피클',
    '맥주' : '땅콩',
    '치킨' : '치킨무',
    '삼겹살' : '파채',
}

while 1:
  a = input("좋아하는 음식은? '떡볶이', '짜장면','라면','피자','맥주','치킨','삼겹살'")

  if a in menu:
      print(f'궁합음식: {a} ---> {menu[a]}')
  elif a == 'q':
      break
  else:
      print('그런 음식은 여기에 없습니다. 다시 입력하세요')
```

```python
#2유형

menu = {
    '떡볶이' : '순대',
    '짜장면' : '짬뽕',
    '라면' : '김치',
    '피자' : '피클',
    '맥주' : '땅콩',
    '치킨' : '치킨무',
    '삼겹살' : '파채',
}

while True:
    name = list(menu.keys())
    a = input("좋아하는 음식은?" + str(name))
    if a in name:
        print(menu[a])
    else:
        print('그런 음식은 여기에 없습니다. 다시 입력하세요')
        break
```

```python
#3유형
menu = {
    '떡볶이' : '순대',
    '짜장면' : '짬뽕',
    '라면' : '김치',
    '피자' : '코카콜라',
    '맥주' : '땅콩',
    '치킨' : '소떡소떡',
    '삼겹살' : '소주',
}

while True:
    want = input("저녁메뉴 고르시오: " + str(list(menu.keys())))
    if want in menu:
        print("궁합에 맞는 음식은 %s ------> %s "%(want,menu.get(want)))
    elif want == "없음":
        break
    else:
        print("메뉴를 다시 고르세요")
```

```python
#실습

place = {
    '부산' : '해운대',
    '울산' : '태화강',
    '광주' : '펭귀마을',
    '서울' : '남산타워',
    '인천' : '영종도',
    '강원도' :'설악산',
    '춘천' : '남이섬',
    '남해' : '다랑이마을',
    '하동' : '녹차밭',
    '대구' : '동성로',
}

while True:
    a = input("가고싶은 지역을 입력하세요!: \n" + str(list(place.keys())))

    if a in place:
        print("%s에 가시면 %s 장소를 추천드려요 "%(a,place.get(a)))
    elif a == "없음":
        break
    else:
        print("입력하신 지역이 없습니다. 다시 입력해주세요~~")
```
