```python
#변수에 문자열로 된 정수와 실수를 저장하면

a = '345' #문자 >> 정수(int: integer)
b = '3.14' #문자 >> 실수(float)

print(a)
print(b)
print(type(a))
print(type(b))
```

```python
#문자 >> 정수(int: integer)

a = int(a)
print(type(a))
```

```python
#문자 >> 실수(float)
b = float(b)  #int로 바꿀시 강제로 정수로 변경해서 3으로 나옴
print(type(b))
```
