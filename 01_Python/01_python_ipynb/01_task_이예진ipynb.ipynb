02-3 똑똑한 chatGPT 일 시키기

```python
for i in range(1, 20):
        for j in range(1, 20):
            print(f"{i} * {j} = {i*j:2d}", end ="\t")
        print()

#end ="\t"은 탭으로 구분하여 다음 값을 출력하도록 저장
#2d는 출력할 숫자가 두 자릿수일 때도 자릿수를 맞춰주기 위한 포맷 지정자이다.
```

Do it 실습)파이썬 파일로 저장하고 실행하기

```python
#19단 곱셈단 좀 더 가지런하게 출력되게 해줘

for i in range(2,20):
    for j in range(1, 20):
        print("{:2d} x {:2d} = {:3d}".format(i,j,i*j), end="")
    print()
```

```python
#한국어로 된 문장만 콤파일 해주는 정규식 코드
import re

text = "This is a sample sentence. 이것은 샘플 문장입니다. Here is another sample sentence. 여기 또 다른 샘플 문장이 있습니다."

korean_sentences = re.findall(r'[ㄱ-ㅎ가-힣]+', text)
for sentence in korean_sentences:
    print(sentence)
```

```python
#파이썬 슬라이싱 설명해줘

'''
파이썬에서의 슬라이싱은 시퀀스(sequence) 타입인 문자열(String), 리스트(List), 튜플(Tuple), 배열(Array) 등의 객체에서 일부분을 추출하는 기능을 말합니다. 이는 시퀀스의 일부를 새로운 시퀀스로 만들어내는데 사용됩니다.

슬라이싱은 다음과 같은 형식을 가집니다:

python
Copy code
sequence[start:stop:step]
여기서:

sequence: 슬라이싱을 수행할 시퀀스 객체입니다. 이는 문자열, 리스트, 튜플 등이 될 수 있습니다.
start: 슬라이스의 시작 인덱스를 나타냅니다. 이 인덱스에 해당하는 요소는 포함됩니다. 기본값은 0입니다.
stop: 슬라이스의 종료 인덱스를 나타냅니다. 이 인덱스에 해당하는 요소는 포함되지 않습니다. 기본값은 시퀀스의 길이입니다.
step: 선택적으로 지정할 수 있는 슬라이스 간격입니다. 기본값은 1입니다.
위의 형식에서 start, stop, step은 모두 생략 가능하며, 생략 시에는 기본값이 사용됩니다.
'''
```

02-4 예외 처리의 중요성

```python
#오류를 처리하는 방법: try-except-finally
try:
    print(1/0) #try : 예외가 발생할 가능성이 있는 코드를 포함
except ZeroDivisionError: # except : 예외가 발생했을 때 실행할 코드를 포함
    print("Division ny zero error") #finally : 예외가 발생하든 발생하지 않든 항상 실행하는 코드를 포함
finally:
    print("This is will always be printed")

```

#어떤 오류가 있나:

```python
#NameError : 이름을 정의하지 않음
'''
print(foo)

NameError: name 'foo' is not defined
'''
```

```python
#TypeError : 잘못된 유형의 값을 사용 / 정수와 문자열 더하면 오류
'''
print(1 + "hello")
TypeError: unsupported operand type(s) for +: 'int' and 'str'
'''
```

```python
#Va;ueError : 잘못된 값을 사용 / 문자를 정수로 변환할려하면 오류

'''
int("hello")
ValueError: invalid literal for int() with base 10: 'hello'
'''
```

```python
#IndexError : 인덱스 범위를 벗어남

'''
my_list = [1,2,3] #인덱스 범위 0-2
print(my_list[3])
IndexError: list index out of range
'''
```

```python
#KeyError : 딕셔너리에 없는 키에 접근하면 오류

'''
my_dict = {}
print(my_dict['key'])
KeyError: 'key'
'''
```

```python
#AttributeError: 개체에 없는 속성에 접근하면 오류

'''
class MyClass:
    def __init__(self):
        self.my_attribute = 42

# MyClass 객체를 생성
my_object.attribute
AttributeError: 'MyClass' object has no attribute 'attribute'
'''
```

```python
#SyntaxError : 잘못된 문법으로 코드를 작성하면 오류

'''
if 1 < 2 :
    print("hello")
else #  :가 누락
    print("world")

SyntaxError: expected ':'
'''

```

```python
#indentationError : 들여쓰기가 올바르지 않으면 생기는 오류

'''
if 1 < 2:
print("hello")
print("world")

IndentationError: expected an indented block after 'if' statement on line 2
'''
```

```python
#ImportError : 모듈을 임포트 할 수 없다. 올바르게 설치가 안됐거나 없는 모듈을 임포트하면 오류 발생
#import my_module
```
