sequence의 indexing 과 slicing
- indexing : sequence에서 한 원소를 가져오는 것
- slicing : sequence에서 일부분을 가져오는 것

```python
my_str = 'impossible'
my_list = ['Apple', 'Banana','Chamae','Durian']
```

```python
my_str[2] # indexing
```

```python
my_list[2:] # slicing
```
