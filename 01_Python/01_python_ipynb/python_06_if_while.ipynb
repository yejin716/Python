```python
#의류 사이즈 분류
#100에서 10단위로 커지는 사이즈에서
#분류기준
#1) size가 150이하 : 아동용
#2) size가 190이상 : 맞춤옷
#3) 그외 : 성인용
```

```python
size = int(input())

while True:
    if size % 10 == 0:
      if size <=150:
        print("아동용")
      elif size >= 190:
        print("맞춤옷")
      else:
        print("성인용")
      break

    else:
      print("10단위의 사이즈를 다시 입력하세요!")
    break

```

```python
#강사님 답
#10단위로 입력하게 해야 진행 가능한 조건

def Input_ten_unit(number):
    return number % 10 == 0
while True:
    try:
        size = int(input("10단위로 증가되게 입력해 주세요: "))
        if not Input_ten_unit(size):
            print("유효한 입력값이 아니예요. 다시 입력해 주세요")
            continue
        if size <=150:
            print("아동용")
        elif size >= 190:
            print("맞춤옷")
        else:
            print("성인용")
        break
    except ValueError:
        print("유효한 값을 넣어주세요.")
```

```python
'''
1. break : while 반복문 밖으로 빠져나감
2. continue : 다음 소스 코드를 실행하지 않고 다음 단계로 넘어감
3. pass : 아무것도 하지 않고 넘김
'''
```
