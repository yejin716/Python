list_comprehension(리스트 컴프리헨션)

```python
#반복문을 사용. 0~10까지 숫자를 리스트로 만들어요

number = []

for i in range(0,11):
    number = i
    print(number)
```

```python
#for문 사용 정답!
number = []

for i in range(10+1):
    number.append(i) #리스트 3줄

print (number)
```

```python
#list comprehension

number = [i for i in range(10+1)] # 1줄로

print(number)
```

```python
number = []
for i in range(0,11):
    number.append(i) #리스트 3줄

print (number)
```

조건문이 있는 리스트 컴프리헨션

```python
#0~10까지 숫자 중에서 홀수만 되어서
odds = []

for i in number(11):
    if i % 2 == 1:
      odds.append(i)

print(odds)
```

```python
#정답

odds = []

for i in range(10+1):
    if i % 2 == 1:
        odds.append(i)

print(odds)
```

```python
#list comprehension

odds = [i for i in range(10+1) if i % 2 == 1]

print(odds)
```

조건이 2개 라면?

```python
# 0~10까지 숫자 중 , 홀수이고 5보다 작은 수

odds = []

for i in range(10 +1):
    if (i % 2 == 1) and (i < 5):
        odds.append(i)

print(odds)
```

```python
odds = [i for i in range(10+1) if i % 2 == 1 if i < 5]
print(odds)
```

이중 반복문(for문이 2개) 를 list comprehension으로 바꾼다면?



```python
가게 = ['올리브영','세븐일레븐','엘지25']
제품 = ['폰 충전기','물티슈','콜라']
```

```python
살거 = [( x,y )for x in 가게 for y in 제품 ]

print(살거)
```
